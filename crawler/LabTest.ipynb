```python
import requests
from bs4 import BeautifulSoup as bs
url = 'https://www.flyingv.cc/type/online'
res = requests.get(url)
#print res.text
```

```python
soup = bs(res.text)
div = soup.select('.portfolio-content')
index = 0
for ele in div[0:-1]:
#print div[ele]
    proName = ele.select('h5')[0]
    name = proName.text.split('<h5>')
    pronominate = ele.select('.unit-owner')[0]
    nominate = pronominate.text.strip().split('<a href>')
    promoney = ele.select('.unit-money')[0]
    money = promoney.text.split('$')
    prodays =  ele.select('.unit-time')[0]
    days = prodays.text.split('</i>')
    index = index +1
    print str(index) + " " +name[0] + " " + nominate[0] + " " + money[1] + " " + days[0].strip()
    print "========================"
```
